We want to have 'Fuzzy C-means' & 'k-means' on below data sets links
Iris:
https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data
https://archive.ics.uci.edu/ml/machine-learning-databases/iris/bezdekIris.data
Ecoli:
https://archive.ics.uci.edu/ml/machine-learning-databases/ecoli/ecoli.data
Heart disease:
https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data

```python
import matplotlib.pyplot as plt
import pandas as pd
from fcmeans import FCM
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
def handle_MissingValues(theDataset):
    for i in range(len(theDataset.columns)):
        theDataset.iloc[:,i].fillna(theDataset.iloc[:,i].mean(), inplace=True)
def delete_DublicateInstances(theDataset):
    theDataset.drop_duplicates(inplace=True)
def select_NumericAttributes(theDataset):
    return theDataset.select_dtypes(include=['int64', 'float64'])
```

1. First we have 'Fuzzy C-means'.

```python
url='https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
irisDataSet=pd.read_csv(url, names=['Sepal length', 'Sepal width', 'Petal length', 'Petal width', 'class'])
irisDataSet_copy=irisDataSet.copy()
irisDataSet_copy.drop('class', axis=1,inplace=True)
delete_DublicateInstances(irisDataSet_copy)
irisDataSet_copy=select_NumericAttributes(irisDataSet_copy)
handle_MissingValues(irisDataSet_copy)
fcm_irisDataSet_copy=FCM(n_clusters=2)
fcm_irisDataSet_copy.fit(irisDataSet_copy.iloc[:,0:2].to_numpy())
plt.scatter(irisDataSet_copy.iloc[:,0],irisDataSet_copy.iloc[:,1], c=fcm_irisDataSet_copy.predict(irisDataSet_copy.iloc[:,0:2].to_numpy()), cmap='coolwarm', marker='*')
```

```python
url='https://archive.ics.uci.edu/ml/machine-learning-databases/iris/bezdekIris.data'
bezdekIrisDataSet=pd.read_csv(url,names=['Sepal length','Sepal width','Petal length','Petal width','class'])
bezdekIrisDataSet_copy=bezdekIrisDataSet.copy()
bezdekIrisDataSet_copy.drop('class',axis=1,inplace=True)
delete_DublicateInstances(bezdekIrisDataSet_copy)
bezdekIrisDataSet_copy=select_NumericAttributes(bezdekIrisDataSet_copy)
handle_MissingValues(bezdekIrisDataSet_copy)
fcm_bezdekIrisDataSet_copy=FCM(n_clusters=2)
fcm_bezdekIrisDataSet_copy.fit(bezdekIrisDataSet_copy.iloc[:,0:2].to_numpy())
plt.scatter(bezdekIrisDataSet_copy.iloc[:,0],bezdekIrisDataSet_copy.iloc[:,1], c=fcm_bezdekIrisDataSet_copy.predict(bezdekIrisDataSet_copy.iloc[:,0:2].to_numpy()), cmap='RdYlGn', marker='*')
```

```python
url='https://archive.ics.uci.edu/ml/machine-learning-databases/ecoli/ecoli.data'
ecoliDataSet=pd.read_csv(url,names=['Sequence Name','mcg','gvh','lip','chg','aac','alm1','alm2','class'],sep='\s+')
ecoliDataSet_copy=ecoliDataSet.copy()
ecoliDataSet_copy.drop('class',axis=1,inplace=True)
delete_DublicateInstances(ecoliDataSet_copy)
ecoliDataSet_copy=select_NumericAttributes(ecoliDataSet_copy)
handle_MissingValues(ecoliDataSet_copy)
fcm_ecoliDataSet_copy=FCM(n_clusters=2)
fcm_ecoliDataSet_copy.fit(ecoliDataSet_copy.iloc[:,0:2].to_numpy())
plt.scatter(ecoliDataSet_copy.iloc[:,0],ecoliDataSet_copy.iloc[:,1], c=fcm_ecoliDataSet_copy.predict(ecoliDataSet_copy.iloc[:,0:2].to_numpy()), cmap='Spectral', marker='*')
```

```python
url='https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data'
processedClevelandDataSet=pd.read_csv(url, names=['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','num'])
processedClevelandDataSet_copy=processedClevelandDataSet.copy()
processedClevelandDataSet_copy.drop('num',axis=1,inplace=True)
delete_DublicateInstances(processedClevelandDataSet_copy)
processedClevelandDataSet_copy=select_NumericAttributes(processedClevelandDataSet_copy)
handle_MissingValues(processedClevelandDataSet_copy)
fcm_processedClevelandDataSet_copy=FCM(n_clusters=2,)
fcm_processedClevelandDataSet_copy.fit(processedClevelandDataSet_copy.iloc[:,0:2].to_numpy())
plt.scatter(processedClevelandDataSet_copy.iloc[:,0],processedClevelandDataSet_copy.iloc[:,1],c=fcm_processedClevelandDataSet_copy.predict(processedClevelandDataSet_copy.iloc[:,0:2].to_numpy()),cmap='bwr',marker='+')
```

2. Second we have 'K-means'.

```python
url='https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
irisDataSet=pd.read_csv(url,names=['Sepal length','Sepal width','Petal length','Petal width','class'])
irisDataSet_copy=irisDataSet.copy()
irisDataSet_copy.drop('class',axis=1,inplace=True)
delete_DublicateInstances(irisDataSet_copy)
irisDataSet_copy=select_NumericAttributes(irisDataSet_copy)
handle_MissingValues(irisDataSet_copy)
kmeans_irisDataSet_copy=KMeans(n_clusters=2, n_init=10)
kmeans_irisDataSet_copy.fit(irisDataSet_copy.iloc[:,0:2])
plt.scatter(irisDataSet_copy.iloc[:,0],irisDataSet_copy.iloc[:,1],c=kmeans_irisDataSet_copy.fit_predict(irisDataSet_copy.iloc[:,0:2]),cmap='coolwarm',marker='*')
```
